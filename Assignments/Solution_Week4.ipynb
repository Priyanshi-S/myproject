## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.cos(x),color="black",linestyle="-.")
plt.ylabel("y",size=20,color="purple",)
plt.xlabel("x",size=20,color="purple")
plt.title("X-Y graph",color="purple",size=20)
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
clothing=np.array([82,78,48,46,39])
equipment=np.array([48,31,51,29,18])
accessories=np.array([15,20,14,13,9])
var=[x for x,_ in enumerate(store)]
width=0.8
plt.bar(var,clothing,width,color="tomato",label="Clothing",bottom=equipment)
plt.bar(var,equipment,width,color="violet",label="Equipment",bottom=accessories)
plt.bar(var,accessories,width,color="cyan",label="Accessories")
plt.ylabel("AMOUNT",size=12)
plt.xlabel("STORES",size=12)
plt.xticks(var,store,rotation=90)
plt.title("STORES COMPLEX",size=12)
plt.grid(False)
plt.legend(loc="upper right")
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
data=sns.load_dataset("titanic")
sns.barplot(x="pclass",hue="class",y="survived",data=data,palette="spring",saturation=0.85,linewidth =2,order=[3,2,1],dodge=False)
plt.xlabel("Passenger Class",size=12)
plt.ylabel("Survival of Passenger",size=12)
plt.title("No of survival in titanic",size=12)
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
tips=sns.load_dataset("tips")
sns.set_style("darkgrid")
sns.distplot(tips["total_bill"],color="green",vertical=True)
plt.ylabel("Total Bill",size=13)
plt.xlabel("Distribution",size=13)
plt.title("Tips Dataset",size=13)
```

# Q.7
- Also plot the KDE Graph for the above data

```python
sns.kdeplot(tips["total_bill"],shade=True,color="green",vertical=True)
plt.ylabel("Total Bill",size=13)
plt.xlabel("Distribution",size=13)
plt.title("Tips Dataset",size=13)
```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
Country=["United States","Great Britain","India","Russia","Germany"]
Medal=[46,27,26,19,17]
color=["aquamarine","coral","orchid","yellowgreen","teal"]
plt.figure(figsize = (8,5))
explode=(0,0,0.2,0,0)
plt.pie(Medal,labels=Country,colors=color,autopct="%1.2f%%",explode=explode,pctdistance=0.7,shadow=True)
plt.title("Medals",size=15)
plt.show()
```

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
