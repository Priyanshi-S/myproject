## Handle categorical variables with numerically coded values

```python
# Changing categorical data into numerical value

# School
print("School")
print(data.school.unique())
data.loc[data['school']=='GP', 'school'] = 1
data.loc[data['school']=='MS', 'school'] = 2
print(data.school.unique())
print()

# Gender
print("Gender")
print(data.sex.unique())
data.loc[data['sex']=='M', 'sex'] = 1
data.loc[data['sex']=='F', 'sex'] = 2
print(data.sex.unique())
print()

# Address
print("Address")
print(data.address.unique())
data.loc[data['address']=='R', 'address'] = 1
data.loc[data['address']=='U', 'address'] = 2
print(data.address.unique())
print()

# Family Size
print("Family Size")
print(data.famsize.unique())
data.loc[data['famsize']=='LE3', 'famsize'] = 1
data.loc[data['famsize']=='GT3', 'famsize'] = 2
print(data.famsize.unique())
print()

# Parent Status
print("Parent Status")
print(data.Pstatus.unique())
data.loc[data['Pstatus']=='A', 'Pstatus'] = 1
data.loc[data['Pstatus']=='T', 'Pstatus'] = 2
print(data.Pstatus.unique())
print()

# Mother's Job Status
print("Mother's Job Status")
print(data.Mjob.unique())
data.loc[data['Mjob']=='teacher', 'Mjob'] = 5
data.loc[data['Mjob']=='services', 'Mjob'] = 4
data.loc[data['Mjob']=='health', 'Mjob'] = 3
data.loc[data['Mjob']=='other', 'Mjob'] = 2
data.loc[data['Mjob']=='at_home', 'Mjob'] = 1
print(data.Mjob.unique())
print()

# Father's Job Status
print("Father's Job Status")
print(data.Fjob.unique())
data.loc[data['Fjob']=='teacher', 'Fjob'] = 5
data.loc[data['Fjob']=='services', 'Fjob'] = 4
data.loc[data['Fjob']=='health', 'Fjob'] = 3
data.loc[data['Fjob']=='other', 'Fjob'] = 2
data.loc[data['Fjob']=='at_home', 'Fjob'] = 1
print(data.Fjob.unique())
print()

# Reasons
print("Reasons")
print(data.reason.unique())
data.loc[data['reason']=='reputation', 'reason'] = 4
data.loc[data['reason']=='course', 'reason'] = 3
data.loc[data['reason']=='home', 'reason'] = 2
data.loc[data['reason']=='other', 'reason'] = 1
print(data.reason.unique())
print()

# Guardians
print("Guardians")
print(data.guardian.unique())
data.loc[data['guardian']=='father', 'guardian'] = 3
data.loc[data['guardian']=='mother', 'guardian'] = 2
data.loc[data['guardian']=='other', 'guardian'] = 1
print(data.guardian.unique())
print()

# School Support
print("School Support")
print(data.schoolsup.unique())
data.loc[data['schoolsup']=='yes', 'schoolsup'] = 2
data.loc[data['schoolsup']=='no', 'schoolsup'] = 1
print(data.schoolsup.unique())
print()

# Family Support
print("Family Support")
print(data.famsup.unique())
data.loc[data['famsup']=='yes', 'famsup'] = 2
data.loc[data['famsup']=='no', 'famsup'] = 1
print(data.famsup.unique())
print()

# Paid
print("Paid")
print(data.paid.unique())
data.loc[data['paid']=='yes', 'paid'] = 2
data.loc[data['paid']=='no', 'paid'] = 1
print(data.paid.unique())
print()

# Activities
print("Activities")
print(data.activities.unique())
data.loc[data['activities']=='yes', 'activities'] = 2
data.loc[data['activities']=='no', 'activities'] = 1
print(data.activities.unique())
print()

# Nursery
print("Nursery")
print(data.nursery.unique())
data.loc[data['nursery']=='yes', 'nursery'] = 2
data.loc[data['nursery']=='no', 'nursery'] = 1
print(data.nursery.unique())
print()

# Higher Education
print("Higher Education")
print(data.higher.unique())
data.loc[data['higher']=='yes', 'higher'] = 2
data.loc[data['higher']=='no', 'higher'] = 1
print(data.higher.unique())
print()

# Internet
print("Internet")
print(data.internet.unique())
data.loc[data['internet']=='yes', 'internet'] = 2
data.loc[data['internet']=='no', 'internet'] = 1
print(data.internet.unique())
print()

# Relationship
print("Relationship")
print(data.romantic.unique())
data.loc[data['romantic']=='yes', 'romantic'] = 2
data.loc[data['romantic']=='no', 'romantic'] = 1
print(data.romantic.unique())
print()
```

## Perform Univariate Analysis

```python
col = ['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences', 'G1', 'G2', 'G3']

fig, ax = plt.subplots(len(col), figsize=(18,50))

for i, col_val in enumerate(col):
    sns.distplot(data[col_val], hist=True, ax=ax[i])
    ax[i].set_title(col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=10)
    ax[i].set_ylabel('Value', fontsize=10)

plt.show()
```

##  Dataset Outliers

```python
col = ['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences', 'G1', 'G2', 'G3']
fig, ax = plt.subplots(len(col), figsize=(34,60))
for i, col_val in enumerate(col):
    sns.boxplot(y=data[col_val], ax=ax[i])
    ax[i].set_title('Box plot - {}'.format(col_val), fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
plt.show()
```
